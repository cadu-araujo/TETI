```python
world = gpd.read_file(path+ 'ne_110m_admin_0_countries.shp')
```

```python
latitude = -3.119028
longitude = -60.021731
city_location = gpd.GeoDataFrame(geometry=[Point(longitude, latitude)], crs="EPSG:4326")
ax = world.plot(figsize=(10, 6))
city_location.plot(ax=ax, color='red', markersize=50)
plt.show()
```

```python
dt_m = pd.read_csv(path+ 'municipios.csv')
```

```python
dt_m.info()
```

```python
def mostra_no_mapa (nome):
    latitude = dt_m[dt_m['nome'] == nome][['latitude']].values[0]
    longitude = dt_m[dt_m['nome'] == nome][['longitude']].values[0]
    city_location = gpd.GeoDataFrame(geometry=[Point(longitude, latitude)], crs="EPSG:4326")
    ax = world.plot(figsize=(10, 6))
    city_location.plot(ax=ax, color='red', markersize=50)
    plt.show()
```

```python
mostra_no_mapa('')
```
